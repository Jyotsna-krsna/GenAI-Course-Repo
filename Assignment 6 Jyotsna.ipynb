```python
#function that finds and replaces all occurrences of a given word in a file named data.txt with another word.

def create_file():
  with open('document.txt','w') as f:
    f.write("Apples are red. Apples are sweet.\n")
    f.write('Apples are healthy.\n')

def read_file():
  with open("document.txt",'r') as f:
    print(f.read())

def count_words():
  with open('document.txt','r') as f:
    words=f.read()
    words=words.split()
    return words


create_file()
print("Original file")
read_file()


print("Number of words in the document")
words=count_words()
print(f'there are {len(words)} words in the file.')
print(f'The words are {words}')


```
