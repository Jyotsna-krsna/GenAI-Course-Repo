```python
#Write a function that appends a given string to the end of a file named log.txt.
def create_file():
  with open('log.txt','w') as f:
    f.write("1.Line 1 \n")
    f.write('2. Line 2 \n')

def appendto_file():
  with open('log.txt','a') as f:
    f.write('This is new line\n')

def read_file():
  with open("log.txt",'r') as f:
    print(f.read())



#call function write_file
create_file()
print("Original file")
read_file()


print("Modified file")
appendto_file()
read_file()


```
