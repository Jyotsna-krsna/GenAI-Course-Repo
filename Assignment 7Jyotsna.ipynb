```python
#a function that reads the contents of a file named reverse.txt and prints each line in reverse order.

def create_file():
  with open('document.txt','w') as f:
    f.write("Apples are red. Apples are sweet.\n")
    f.write('Apples are healthy.\n')

def read_file():
  with open("document.txt",'r') as f:
    print(f.read())

def read_lines():
  with open('document.txt','r') as f:
    lines=f.readlines()
    for line in lines.reversed():
      print(line)


create_file()
print("Original file")
read_file()


print("Lines")
read_lines()


```
